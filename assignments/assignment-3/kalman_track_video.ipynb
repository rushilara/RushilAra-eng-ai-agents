# Kalman drone tracking (filterpy)

Reads per-frame JSON from Task 1 (`{video_stem}_frame000000.json`, …), runs a constant-velocity Kalman filter on the bbox center, and writes one `{stem}_tracked.mp4` per input video with detector box + trajectory polyline.

Requires: `filterpy`, OpenCV, ffmpeg on PATH. Install: `pip install -r requirements-rtdetr.txt`.

```python
from pathlib import Path

from kalman_track_video import process_one_video

HW3 = Path(".").resolve()
videos_dir = HW3 / "videos"
detections_dir = HW3 / "videos" / "detections"
out_dir = HW3 / "videos" / "tracked"

video_path = videos_dir / "drone_video_1.mp4"
if not video_path.is_file():
    raise FileNotFoundError(f"Place {video_path.name} under {videos_dir}")

out_path = out_dir / f"{video_path.stem}_tracked.mp4"
n = process_one_video(video_path, detections_dir, out_path, max_missed=8, decoder="auto")
print(out_path, n, "frames")
```

```python
# Batch all MP4s in videos_dir (same as CLI)
from kalman_track_video import main as cli_main
import sys

sys.argv = [
    "kalman_track_video.py",
    "--videos-dir", str(videos_dir),
    "--detections-dir", str(detections_dir),
    "--output-dir", str(out_dir),
    "--decoder", "auto",
]
cli_main()
```
