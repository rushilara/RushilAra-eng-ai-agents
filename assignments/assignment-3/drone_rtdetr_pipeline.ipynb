# Drone detection: RT-DETR fine-tune + video JSON

2. Run **Prepare** (symlinks + `data/yolo_ultralytics/drone.yaml`), then **Train** (GPU recommended), then **Video** on a folder of `.mp4` files.

Equivalent CLI: `python prepare_yolo_dataset.py`, `python train_rtdetr.py`, `python run_video_inference.py --help`.

```python
from pathlib import Path
import subprocess
import sys

HW3 = Path('.').resolve()
subprocess.run([sys.executable, str(HW3 / 'prepare_yolo_dataset.py')], check=True, cwd=HW3)
```

```python
# Fine-tune RT-DETR-L (adjust epochs; requires GPU for reasonable time)
import subprocess
import sys
from pathlib import Path

HW3 = Path('.').resolve()
cmd = [
    sys.executable, str(HW3 / 'train_rtdetr.py'),
    '--epochs', '200',
    '--patience', '30',
    '--weight-decay', '0.0005',
    '--val-test-after',
]
subprocess.run(cmd, check=True, cwd=HW3)
```

```python
# Per-frame JSON under videos_dir/detections/ — set WEIGHTS to your best.pt
import subprocess
import sys
from pathlib import Path

HW3 = Path('.').resolve()
WEIGHTS = HW3 / 'runs/rtdetr/drone_finetune/weights/best.pt'
VIDEOS = HW3 / 'videos'  # put .mp4 files here or change path

cmd = [
    sys.executable, str(HW3 / 'run_video_inference.py'),
    '--videos-dir', str(VIDEOS),
    '--weights', str(WEIGHTS),
    '--conf', '0.25',
]
subprocess.run(cmd, check=True, cwd=HW3)
```
